```python
import NMF as nmf
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2
```

```python
seed = 1998
np.random.seed(seed)
```

```python
V = np.abs(np.random.randn(10, 25))
fac_V = nmf.NMF(V)
```

## EM IS

```python
# abs(N(0,1) + 1) random initialisation
n_iter = 10
costs_iter = []
for i in range(n_iter):
    W, H, WH = fac_V.factorize_EM_IS(5, 5000, threshold=1E-16)
    costs_iter.append(np.copy(np.array(fac_V.costs)))    
```

```python
len_costs = costs_iter[0].shape[0]

fig = plt.figure()
ax = fig.add_subplot(111, title=r'costs for default initialisation')
for costs in costs_iter:
    ax.semilogy(range(len_costs), costs)
ax.set_xlabel('iteration')
ax.set_ylabel('cost')
plt.show()
```

```python
#  abs(N(0,1)) random initialisation
n_iter = 10
costs_iter = []
for i in range(n_iter):
    W, H, WH = fac_V.factorize_EM_IS(5, 5000, threshold=1E-16, W_0=np.abs(np.random.randn(10, 5)), H_0=np.abs(np.random.randn(5, 25)))
    costs_iter.append(np.copy(np.array(fac_V.costs)))
```

```python
len_costs = costs_iter[0].shape[0]

fig = plt.figure()
ax = fig.add_subplot(111, title=r'costs for abs(N(0,1))')
for costs in costs_iter:
    ax.semilogy(range(len_costs), costs)
ax.set_xlabel('iteration')
ax.set_ylabel('cost')
plt.show()
```

```python
#  abs(N(0,1)) (W) amplitude*abs(N(0,1)) (H) random initialisation
amplitude = 10
n_iter = 10
costs_iter = []
for i in range(n_iter):
    W, H, WH = fac_V.factorize_EM_IS(5, 5000, threshold=1E-16, W_0=np.abs(np.random.randn(10, 5)), H_0=amplitude*np.abs(np.random.randn(5, 25)))
    costs_iter.append(np.copy(np.array(fac_V.costs)))
```

```python
len_costs = costs_iter[0].shape[0]

fig = plt.figure()
ax = fig.add_subplot(111, title=r'costs for (abs(N(0,1)), amplitude$\cdot$abs(N(0,1)))')
for costs in costs_iter:
    ax.semilogy(range(len_costs), costs)
ax.set_xlabel('iteration')
ax.set_ylabel('cost')
plt.show()
```

```python
#  abs(N(0,1)) (W) abs(amplitude*N(0,1)) + baseline (H)) random initialisation
amplitude = 10
baseline = 100
n_iter = 10
costs_iter = []
for i in range(n_iter):
    W, H, WH = fac_V.factorize_EM_IS(5, 5000, threshold=1E-16, W_0=np.abs(np.random.randn(10, 5)), H_0=np.abs(amplitude*np.random.randn(5, 25) + baseline))
    costs_iter.append(np.copy(np.array(fac_V.costs)))
```

```python
len_costs = costs_iter[0].shape[0]

fig = plt.figure()
ax = fig.add_subplot(111, title=r'costs for (abs(N(0,1)), abs(amplitude$\cdot$N(0,1)$+$baseline)')
for costs in costs_iter:
    ax.semilogy(range(len_costs), costs)
ax.set_xlabel('iteration')
ax.set_ylabel('cost')
plt.show()
```
