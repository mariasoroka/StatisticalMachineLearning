# Convergence of NMF

```python
V = np.abs(np.random.randn(10, 25))
fac_V = nmf.NMF(V)
```

## EM IS

```python
# abs(N(0,1) + 1) random initialisation
n_iter = 10
costs_iter = []
for i in range(n_iter):
    W, H, WH = fac_V.factorize_EM_IS(5, 5000, threshold=1E-16)
    costs_iter.append(np.copy(np.array(fac_V.costs)))    
```

```python
len_costs = costs_iter[0].shape[0]

fig = plt.figure()
ax = fig.add_subplot(111, title=r'costs for default initialisation')
for costs in costs_iter:
    ax.semilogy(range(len_costs), costs)
ax.set_xlabel('iteration')
ax.set_ylabel('cost')
plt.show()
```

# Synthethic Data

```python
K = 2
F = 45
N = 55

W = np.zeros((F, K))
for k in range(K):
    W[:, k] = (np.array(range(F))**(k+1)+1)+10
    
H = np.zeros((K, N))
for k in range(K):
    H[k, :] = (k+1)*(N - np.array(range(N)))+10

V = W @ H

# normalise column of W
for k in range(K):
    norm_factor = np.linalg.norm(W[:, k])
    W[:, k] = W[:, k] / norm_factor
    H[k, :] = H[k, :] * norm_factor

# plot
fig, axes = plt.subplots(K, 2, figsize=(8 * 2, K * 4))
for k in range(K):
    axes[k, 0].plot(range(F), W[:, k])
    axes[k, 1].plot(range(N), H[k, :])
plt.show()
```

```python
fac_V = nmf.NMF(V)
```

## EM IS

```python
#  abs(N(0,1)) (W) abs(amplitude*N(0,1)) + baseline (H)) random initialisation
amplitude = 10
baseline = 100
_W, _H, _WH = fac_V.factorize_EM_IS(K, 5000, threshold=1E-16, W_0=np.abs(np.random.randn(F, K)), H_0=np.abs(amplitude*np.random.randn(K, N) + baseline))
costs = np.copy(np.array(fac_V.costs))

fig = plt.figure()
ax = fig.add_subplot(111, title=r'costs for (abs(N(0,1)), abs(amplitude$\cdot$N(0,1)$+$baseline)')
ax.semilogy(range(len(costs)), costs)
ax.set_xlabel('iteration')
ax.set_ylabel('cost')
plt.show()
```

```python
# plot
fig, axes = plt.subplots(K, 2, figsize=(8 * 2, K * 4))
for k in range(K):
    axes[k, 0].plot(range(F), _W[:, k])
    axes[k, 1].plot(range(N), _H[k, :])
plt.show()
```

## Gaussian Data

```python
import scipy.stats as stats

K = 4
F = 1025
N = 500
W = np.zeros((F, K))
H = np.zeros((K, N))

pitches = np.array([61, 65, 68, 72])
times_set = np.array([[20, 120, 220, 320],[200, 300, 400, 500],[200, 250, 300, 350],[400, 450, 500, 550]])
freq_set = 440 * np.power(2, (pitches - 69) / 12)

frequencies = np.linspace(0, 2000, F)
times = np.linspace(0, 600, N)

for i in range(4):
    W[:, i] = stats.norm.pdf(frequencies, loc=freq_set[i], scale=5) + 0.5 * stats.norm.pdf(frequencies, loc=2 * freq_set[i], scale=5) + 100
    for j in range(4):
        H[i, :] += (times > times_set[i, j]) * stats.norm.pdf(times, loc=times_set[i, j], scale=10) + 100
        
for k in range(K):
    norm_factor = np.linalg.norm(W[:, k])
    W[:, k] = W[:, k] / norm_factor
    H[k, :] = H[k, :] * norm_factor

fig, axes = plt.subplots(K, 2, figsize=(8 * 2, K * 4))
for i in range(K):
    axes[i, 0].plot(frequencies, W[:, i])
    axes[i, 1].plot(times, H[i, :])
plt.show()
```

```python
fac_V = nmf.NMF(W @ H)
```

```python
#  abs(N(0,1)) (W) abs(amplitude*N(0,1)) + baseline (H)) random initialisation
amplitude = 10
baseline = 100
_W, _H, _WH = fac_V.factorize_EM_IS(K, 5000, threshold=1E-16, W_0=np.abs(np.random.randn(F, K)), H_0=np.abs(amplitude*np.random.randn(K, N) + baseline))
costs = np.copy(np.array(fac_V.costs))

fig = plt.figure()
ax = fig.add_subplot(111, title=r'costs for (abs(N(0,1)), abs(amplitude$\cdot$N(0,1)$+$baseline)')
ax.semilogy(range(len(costs)), costs)
ax.set_xlabel('iteration')
ax.set_ylabel('cost')
plt.show()
```

```python
# plot
fig, axes = plt.subplots(K, 2, figsize=(8 * 2, K * 4))
for k in range(K):
    axes[k, 0].plot(range(F), _W[:, k])
    axes[k, 1].plot(range(N), _H[k, :])
plt.show()
```
