```python
from NMF import *
from sklearn.decomposition import NMF
import numpy as np
%matplotlib inline
```

```python
filename = './data/Chords.wav'
ViolinRecording = Recording(filename)
ViolinSpectrogram = ViolinRecording.compute_spectrogram()
ViolinSpectrogram.visualize()
```

**KL_NMF from the lib**

```python
K = 5
model = NMF(n_components=K, init='random', solver='mu', beta_loss='kullback-leibler', max_iter=5000, tol=1e-6, random_state=0, verbose=0)
W = model.fit_transform(ViolinSpectrogram.compute_matrix())
H = model.components_
for k in range(K):
    norm_factor = np.linalg.norm(W[:, k])
    W[:, k] = W[:, k] / norm_factor
    H[k, :] = H[k, :] * norm_factor
plot_freq_times(W, H, ViolinSpectrogram)
print(model.n_iter_)
```

1. Again 53, but noisy time plot
2. Pitch 61
3. 67.6
4. 68
5. 72.8? Time code corresponds to 61! Split 61 but in different manner.
6. Noise

So, 65 and 72 not detected. 61 and 68 are split.
