```python
from NMF import *
```

```python
filename = './data/Chords.wav'
recording = Recording(filename) # loading recording in .wav format
spectrogram = recording.compute_spectrogram() # computing spectrogram
spectrogram.visualize() # plotting spectrogram values
```

```python
K = 6 # setting number of components for factorisation
matrix_to_factorize = NMF(spectrogram.compute_matrix()) 
W, H, WH = matrix_to_factorize.factorize_KL(K, 5000) # factorising
plot_freq_times(W, H, spectrogram) # plotting obtained components
```

```python
new_spectrogram = Spectrogram(np.sqrt(W @ H), spectrogram.fs) # creating new spectrogram from factorisation
new_spectrogram.visualize() # visualizing new spectrogram
new_spectrogram.restore_recording("./data/reconstructed.wav") # restoring audio from new spectrogram in given file
```
