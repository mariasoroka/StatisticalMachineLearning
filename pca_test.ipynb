```python
filename = './data/test2.wav'
ViolinRecording = Recording(filename)
ViolinSpectrogram = ViolinRecording.compute_spectrogram()
ViolinSpectrogram.visualize()
ViolinSpectrogram.restore_recording("./data/exp.wav")
matrix_to_factorize = ViolinSpectrogram.compute_matrix()
```

```python
K=1
W, H, WH = factorize_PCA(matrix_to_factorize, K, nonnegative = True)
new_spectrogram = Spectrogram(np.sqrt(WH), ViolinSpectrogram.fs)
new_spectrogram.visualize()
new_spectrogram.restore_recording("./data/exp_full_pca.wav")
#PCA also selects notes if K < number of notes
```

```python
W, H, WH = factorize_TruncatedSVD(matrix_to_factorize, K, 1000, nonnegative = True)
new_spectrogram = Spectrogram(np.sqrt(WH), ViolinSpectrogram.fs)
new_spectrogram.visualize()
new_spectrogram.restore_recording("./data/exp_full_svd.wav")
```
